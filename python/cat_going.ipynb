```python
import pandas as pd
import numpy as np
import datetime

event_file = "../out/events.csv"

event_df = pd.read_csv(event_file, names=['time', 'event'])
event_df['time']= pd.to_datetime(event_df['time'],unit='s', utc=True).dt.tz_convert('America/Los_Angeles')
event_df['time_naive']= event_df['time'].dt.tz_localize(None)
event_df['event']= event_df['event'].map({'ENTER': 1, 'EXIT': -1})

print(event_df)
```

```python
import plotly.express as px

fig = px.area(event_df, x='time_naive', y='event', line_shape='hv')
fig.show()
```

```python

sensor_file = "../out/sensor.csv"
sensor_df = pd.read_csv(sensor_file, names=['date', 'time', 'val'])

time_monotonic = [0]
for i in range(1, len(sensor_df)):
    if sensor_df['time'][i] > sensor_df['time'][i-1]:
        delta = sensor_df['time'][i] - sensor_df['time'][i-1]
    else:
        delta = sensor_df['date'][i] - sensor_df['date'][i-1]
    time_monotonic.append(time_monotonic[i-1] + delta)

sensor_df['time_monotonic'] = time_monotonic
```

```python
import plotly.express as px

fig = px.scatter(sensor_df, x='time_monotonic', y='val')
fig.show()
```
