```python
import pandas as pd
import numpy as np
import datetime

event_file = "../out/events.csv"

event_df = pd.read_csv(event_file, names=['time', 'event'])
event_df['time']= pd.to_datetime(event_df['time'],unit='s', utc=True).dt.tz_convert('America/Los_Angeles')
event_df['time_naive']= event_df['time'].dt.tz_localize(None)
event_df['event']= event_df['event'].map({'ENTER': 1, 'EXIT': -1})

print(event_df)
```

```python

periods = {'time':[], 'event':[], 'duration':[]}
for i in range(1, len(event_df)):
    delta = event_df['time'][i] - event_df['time'][i-1]
    periods['duration'].append(delta)
    periods['time'].append(event_df['time'][i-1] + delta/2)
    if event_df['event'][i-1] == 1 and event_df['event'][i] == -1:
        periods['event'].append('INSIDE')
    elif event_df['event'][i-1] == -1 and event_df['event'][i] == 1:
        periods['event'].append('OUTSIDE')
    else:
        periods['event'].append('ERROR')
    
period_df = pd.DataFrame(periods)

print(period_df)
```

```python
import plotly.express as px

fig = px.area(event_df, x='time_naive', y='event', line_shape='hv')
fig.show()
```

```python
dates = period_df['time'].dt.date.unique()

for date in dates:
    mask = period_df['time'].dt.date == date
    df = period_df[mask]
    df_inside = df[df['event'] == 'INSIDE']
    df_outside = df[df['event'] == 'OUTSIDE']
    # print(df)
    print(date)
    print(f"\t Errors: {sum(df['event'] == 'ERROR')}")

    first_exit = df_outside.iloc[0]['time'] - df_outside.iloc[0]['duration']
    last_entry = df_outside.iloc[-1]['time'] + df_outside.iloc[-1]['duration']

    df_inside = df_inside[(df_inside['time'] >= first_exit) & (df_inside['time'] <= last_entry)]
    df_outside = df_outside[(df_outside['time'] >= first_exit) & (df_outside['time'] <= last_entry)]
    print(f"\t First Exit: {first_exit.time()}")
    print(f"\t Last Enter: {last_entry.time()}")
    print(f"\t Inside:")
    print(f"\t\t Count: {len(df_inside['duration'])}")
    print(f"\t\t Total: {df_inside['duration'].sum()} hours")
    print(f"\t\t Min: {df_inside['duration'].min()}")
    print(f"\t\t Max: {df_inside['duration'].max()}")
    print(f"\t\t Mean: {df_inside['duration'].mean()}")

    print(f"\t Outside:")
    print(f"\t\t Count: {len(df_outside['duration'])}")
    print(f"\t\t Total: {df_outside['duration'].sum()} hours")
    print(f"\t\t Min: {df_outside['duration'].min()}")
    print(f"\t\t Max: {df_outside['duration'].max()}")
    print(f"\t\t Mean: {df_outside['duration'].mean()}")
```

```python
sensor_file = "../out/sensor.csv"
sensor_df = pd.read_csv(sensor_file, names=['date', 'time', 'val'])

time_monotonic = [0]
for i in range(1, len(sensor_df)):
    if sensor_df['time'][i] > sensor_df['time'][i-1]:
        delta = sensor_df['time'][i] - sensor_df['time'][i-1]
    else:
        delta = sensor_df['date'][i] - sensor_df['date'][i-1]
    time_monotonic.append(time_monotonic[i-1] + delta)

sensor_df['time_monotonic'] = time_monotonic
sensor_df['time_monotonic'] /= 1000.0
```

```python
import plotly.express as px

df = sensor_df[(sensor_df['time_monotonic'] > 268505) & (sensor_df['time_monotonic'] < 268508)]
df['time_monotonic'] -= df['time_monotonic'].iloc[0] 
fig = px.scatter(df, x='time_monotonic', y='val')
fig.show()
```
